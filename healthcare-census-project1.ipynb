# Problem Statement 5: (Find and process Missing Data)

```python
census_df5 = census_df4.copy()
```

```python
census_df5.head()
```

```python
percent_missing = (census_df5.isna().sum()/ len(census_df5) * 100)
```

```python
percent_missing
```

```python
population_index = census_df5.loc[census_df5['Population'].isna()].index
```

```python
population_index
```

```python
# Population = Male + Female
census_df5.loc[population_index,["Population"]] = census_df5.loc[population_index,"Male"] + census_df5.loc[population_index,"Female"]
```

```python
# male
male_index = census_df5.loc[census_df5['Male'].isna()].index
```

```python
census_df5.loc[male_index,["Male"]] = census_df5.loc[male_index,"Population"] - census_df5.loc[male_index,"Female"]
```

```python
# female
female_index = census_df5.loc[census_df5['Female'].isna()].index
```

```python
census_df5.loc[female_index,["Female"]] = census_df5.loc[female_index,"Population"] - census_df5.loc[female_index,"Male"]
```

```python
# Literate = Literate_Male + Literate_Female
literate_index = census_df5.loc[census_df5['Literate'].isna()].index
```

```python
census_df5.loc[literate_index,["Literate"]] = census_df5.loc[literate_index,"Literate_Male"] + census_df5.loc[literate_index,"Literate_Female"]
```

```python
census_df5[census_df5["Literate"].isna()]
```

```python
# Literate_Male
Literate_Male_index = census_df5.loc[census_df5['Literate_Male'].isna()].index
```

```python
census_df5.loc[Literate_Male_index,["Literate_Male"]] = census_df5.loc[Literate_Male_index,"Literate"] - census_df5.loc[Literate_Male_index,"Literate_Female"]
```

```python
# Literate_Female
Literate_Female_index = census_df5.loc[census_df5['Literate_Female'].isna()].index
```

```python
census_df5.loc[Literate_Female_index,["Literate_Female"]] = census_df5.loc[Literate_Female_index,"Literate"] - census_df5.loc[Literate_Female_index,"Literate_Male"]
```

```python
# Households = Households_Rural + Households_Urban 
households_index = census_df5.loc[census_df5['Households'].isna()].index
```

```python
census_df5.loc[households_index,["Households"]] = census_df5.loc[households_index,"Households_Rural"] + census_df5.loc[households_index,"Urban_Households"]
```

```python
# Households_Rural
Households_Rural_index = census_df5.loc[census_df5['Households_Rural'].isna()].index
```

```python
census_df5.loc[Households_Rural_index,["Households_Rural"]] = census_df5.loc[Households_Rural_index,"Households"] - census_df5.loc[Households_Rural_index,"Urban_Households"]
```

```python
# Households_Urban 
Urban_Households_index = census_df5.loc[census_df5['Urban_Households'].isna()].index
```

```python
census_df5.loc[Urban_Households_index,["Urban_Households"]] = census_df5.loc[Urban_Households_index,"Households"] - census_df5.loc[Urban_Households_index,"Households_Rural"]
```

```python
census_df5.isna().sum()
```

```python
# Young_and_Adult, Middle_Aged, Senior_Citizen ,Age_Not_Stated

census_df6['Young_and_Adult'].fillna(census_df6['Population'] - (census_df6['Middle_Aged'] + census_df6['Senior_Citizen'] + census_df6['Age_Not_Stated']), inplace=True)

census_df6['Middle_Aged'].fillna(census_df6['Population'] - (census_df6['Young_and_Adult'] + census_df6['Senior_Citizen'] + census_df6['Age_Not_Stated']), inplace=True)

census_df6['Senior_Citizen'].fillna(census_df6['Population'] - (census_df6['Young_and_Adult'] + census_df6['Middle_Aged'] + census_df6['Age_Not_Stated']), inplace=True)

census_df6['Age_Not_Stated'].fillna(census_df6['Population'] - (census_df6['Young_and_Adult'] + census_df6['Senior_Citizen'] + census_df6['Middle_Aged']), inplace=True)
```

```python
def Alter(hospitals):
#    replacing laddakh with jammu and kashmir
    census_updated =census_df7['State/UT'].replace(['laddakh'],['jammu and kashmir'])
    housing_updated=housing_df['State/UT'].replace(['laddakh'],['jammu and kashmir'])
    hospitals['State/UT']=hospitals['State/UT'].str.lower().str.replace('*','')

    cencus_states=sorted(census_updated.unique().tolist())
    
    housing_states=sorted(housing_updated.unique().tolist())
    hospital_states=sorted(hospitals['State/UT'].unique().tolist())
    #Checking for the number of states of in each of the datasets
    print(len(cencus_states),len(housing_states),len(hospital_states))
    states_with_diff_names=[]
    for i in range(1,len(cencus_states)):
        if cencus_states[i-1]!=housing_states[i-1] or cencus_states[i-1]!=hospital_states[i] or housing_states[i-1]!=hospital_states[i]:
            c,hou,hos=cencus_states[i],housing_states[i],hospital_states[i]
            states_with_diff_names.append((c,hou,hos))
    print(states_with_diff_names)
    hospitals['State/UT']=hospitals['State/UT'].str.replace('&','and')
    hospitals['State/UT']=hospitals['State/UT'].replace(['india','odisha','puducherry','delhi'],['all india','orissa','pondicherry','nct of delhi'])
    return hospitals['State/UT']
    
```
